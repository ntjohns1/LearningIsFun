```python
import os
import json
import csv
from pymongo import MongoClient

# Path to the main directory containing subfolders
directory = input("enter a pathname")

# Connect to MongoDB (adjust the connection string as necessary)
client = MongoClient('mongodb://localhost:27017/moneybot')
db = client['moneybot']
collection1 = db['fh_sec_filings_2010_2020']
collection2 = db['fh_sec_filings_1994_2020']

# Iterate through each subdirectory and file
for subdir, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(subdir, file)
            with open(file_path, 'r') as f:
                file_data = json.load(f)
                # Insert the data into MongoDB
                collection1.insert_one(file_data)
                print(f'Inserted file {file}')
        elif file.endswith('.csv'):
            file_path = os.path.join(subdir, file)
            with open(file_path, 'r') as f:
                reader = csv.DictReader(f)
                for row in reader:
                    # Insert the data into MongoDB
                    collection2.insert_one(row)
                print(f'Inserted file {file}')

print("All files have been processed and inserted into MongoDB.")
```
